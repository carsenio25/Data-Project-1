```python
def get_dataframe(user_id, pwd, host_name, db_name, sql_query):
    conn_str = f"mysql+pymysql://{user_id}:{pwd}@{host_name}/{db_name}"
    sqlEngine = create_engine(conn_str, pool_recycle=3600)
    connection = sqlEngine.connect()
    dframe = pd.read_sql(sql_query, connection);
    connection.close()
    
    return dframe


def set_dataframe(user_id, pwd, host_name, db_name, df, table_name, pk_column, db_operation):
    conn_str = f"mysql+pymysql://{user_id}:{pwd}@{host_name}/{db_name}"
    sqlEngine = create_engine(conn_str, pool_recycle=3600)
    connection = sqlEngine.connect()
    
    if db_operation == "insert":
        df.to_sql(table_name, con=connection, index=False, if_exists='replace')
        connection.execute(f"ALTER TABLE {table_name} ADD PRIMARY KEY ({pk_column});")
            
    elif db_operation == "update":
        df.to_sql(table_name, con=connection, index=False, if_exists='append')
    
    connection.close()
```

```python
with open('chinook_tracks.txt', 'r', encoding='utf-8') as file:
    # Iterate through each track name in the file
    for track_name in file:
        # Remove any leading/trailing whitespace from the track name
        track_name = track_name.strip()
        print(track_name)
        # Make a GET request to the Spotify Search API to search for a track by name
        # Concatenating the base URL with the track name for the query and specify search type as 'track'
        user_tracks_response = requests.get("https://api.spotify.com/v1/search?q=track:" + track_name + "&type=track",
                                            params=params, headers=headers)
        # Parse the JSON response from the API call
        data = user_tracks_response.json()
        
        # Check if the response contains track items
        if 'tracks' in data and 'items' in data['tracks'] and len(data['tracks']['items']) > 0:
            # If there were tracks found, access the first track's data
            track_data = data['tracks']['items'][0]
            # Extract the Spotify ID of the first track
            track_id = track_data['id']
            # Add the track name and its corresponding Spotify ID to a dictionary
            id_dict[track_name] = track_id
            print(track_id)
        # Introduce a short time delay in the loop to avoid hitting the Spotify API rate limit
        time.sleep(0.1)
```

```python
track_details = {}
# Iterating over each track name in the id_dict dictionary
for track_name in id_dict:
    # Retrieve the Spotify track ID for the current track name
    spotify_track_id = id_dict[track_name]
    if spotify_track_id is None:
        track_details[track_name] = [None] * 10
        continue
    url = f"https://api.spotify.com/v1/tracks/{spotify_track_id}"
    # Make the API request to Spotify to get info about the track
    track_response = requests.get(url, headers=headers)
    # Parse the JSON response to extract the track details
    track_info = track_response.json()

    #Extracting info about track
    album_id = track_info['album']['id']
    total_tracks_in_album = track_info['album']['total_tracks']
    release_date = track_info['album']['release_date']
    album_name = track_info['album']['name']
    primary_artist_name = track_info['artists'][0]['name']
    primary_artist_id = track_info['artists'][0]['id']
    song_name = track_info['name']
    track_number = track_info['track_number']
    duration_ms = track_info['duration_ms']
    
    # Store extracted song info in dictionary 
    track_details[track_name] = [spotify_track_id, song_name, primary_artist_id, primary_artist_name, release_date, track_number,
                         duration_ms, album_id, album_name,
                         total_tracks_in_album]
    print(track_response)
    print(track_details)
```

```python
import json

formatted_data = {}
for key, values in track_details.items():
    print(values)
    # Format the current item's data and assign it to the corresponding key  
    formatted_data[key] = {
        "spotify_track_id": values[0],
        "song_name": values[1],
        "artist_id": values[2],
        "artist_name": values[3],
        "date": values[4],
        "track_number": values[5],
        "duration_ms": values[6],
        "album_id": values[7],
        "album_name": values[8],
        "track_count": values[9]
    }

# Convert the formatted data dictionary to a JSON string
data_json = json.dumps(formatted_data, indent=4)
print(data_json)

with open("spotify_data.json", "w") as json_file:
    # Writing the formatted data dictionary to the file as JSON
    json.dump(formatted_data, json_file, indent=3)
```
